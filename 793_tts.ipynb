## English Version

```python
from TTS.api import TTS

model_name = "tts_models/multilingual/multi-dataset/xtts_v2"

tts = TTS(model_name=model_name, progress_bar=True)

text = "Hello, this is the presentation of Class 793 final project."

input_voice = "/content/drive/MyDrive/train_voice/sample_voice.wav"
output_path = "demo/en.wav"

tts.tts_to_file(
    text=text,
    speaker_wav=input_voice,
    language='en',
    file_path=output_path,
    split_sentences=False
)
print("Speech synthesis completed. Saved as en.wav")
```

## Chinese Version

```python
text = "你好，这是李牧融的期末项目。"

output_path = "demo/cn.wav"

tts.tts_to_file(
    text=text,
    speaker_wav=input_voice,
    language='zh-cn',
    file_path=output_path,
    split_sentences=False
)
print("语音合成完成，保存为 cn.wav")
```
